```python
total = df1.isnull().sum().sort_values(ascending=False)
miss_val=df1.isnull().sum()/df1.isnull().count()*100
miss_val2=(round(miss_val,1)).sort_values(ascending=False)
missing_data=pd.concat([total,miss_val2],axis=1)
missing_data.columns=['total','%']
missing_data.head(5)
```

```python
total = df2.isnull().sum().sort_values(ascending=False)
miss_val=df2.isnull().sum()/df2.isnull().count()*100
miss_val2=(round(miss_val,1)).sort_values(ascending=False)
missing_data=pd.concat([total,miss_val2],axis=1)
missing_data.columns=['total','%']
missing_data.head(5)
```

```python
df1['price'].replace('\$', '', regex=True, inplace=True)
```

```python
# Remove non-numeric characters from the "price" column
df1['price'] = df1['price'].str.replace('[^\d.]', '', regex=True)

# Convert the "price" column to numeric data type
df1['price'] = pd.to_numeric(df1['price'], errors='coerce')

# Fill missing values in the "price" column with 0
df1['price'] = df1['price'].fillna(0)
```
